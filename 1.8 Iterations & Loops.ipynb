You can iterate over many different object types with a for loop:

```python
# string
for c in "See you on the vertical!":
    print c

```

```python
# list
for e in [1, 2, 3]:
    print("List element", e)
```

```python
# tuple
for e in (1, 2, 3):
    print("Tuple element", e)
```

```python
# Dictionaries
mydict = {'age': 47, 'height': 180}

# dictionary keys
for k in mydict:
    print("Dictionary key", k)
    
# values
for v in mydict.values():
    print("Dictionary value", v)

# key / value pairs (tuples)
for t in mydict.items():
    print("Dictionary key/value pair:", t)
```

```python
# File
with open('/etc/resolv.conf') as f:
    for line in f:
        print(line.rstrip())
```

### Libraries for concurrent and/or parallel execution

- [concurrent.futures](https://docs.python.org/3.7/library/concurrent.futures.html) `ThreadPoolExecutor, ProcessPoolExecutor`
    - Process-based generally best for CPU-bound tasks
    - Thread-based generally best for IO-bound tasks
- [asyncio](https://docs.python.org/3/library/asyncio.html)
    - Framework useful for IO-bound network-related tasks (ex: web servers)

# Exercise: Profiling with cProfile

Profile a script of your choosing using the following command line:  `python -m cProfile my_script.py`

Sample script:

    import time
    
    
    def f1(n: int=3) -> int:
        print("Running SLOW function")
        time.sleep(n)
        return 42
    
    
    def f2() -> int:
        print("Running FAST function")
        return 43
        
    f1()
    f1(5)
    f1(10)
    f2()
    f2()

- sort by cumtime (use the `-s cumtime` option for this)
- save the result to a text file
- save the result in the default binary output format (`-o my.profile`)
- data in the binary format can be visualised using several tools, ex: gprof2dot (needs the graphviz package, which provides the dot command, to be installed)

        gprof2dot -f pstats cProfile.data | dot -Tpng -o my_profile.png
